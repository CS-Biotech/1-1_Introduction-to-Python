### If statements
You now know that expressions can evaluate to True or False, sometimes involving logical and comparison operators, but how do we control what actions get performed in certain situations?

**If statements** control the logical flow of your code. Consider the following pseudocode (describing code but in plain English):
```
if condition is true
    do something
```

You can also do something else if the condition is not true:
```
if condition is true
    do something
otherwise
    do something else
```

You might have multiple different conditions that are mutually exclusive:
```
if condition A is true
    do something
if condition A is not true, but condition B is true
    do something else
if condition A and B are not true, but condition C is true
    do something else
    ...
otherwise
    do something else
```

The above situation shows 3 different conditions (A, B, C), but you can have as many conditions as you want. A few rules:
1. The first condition should be marked by ```if```.
2. All other conditions that follow (if there are any) should be marked by ```elif``` (think of it as "else if")
3. An ```else``` statement at the end (if there is one) indicates what should be done if none of the if/elif statements were True. (Think of it as a default)
4. You must indent the action(s) under each condition.

The program evaluates each condition from the top until one condition evaluates to True, does the actions in that block, then **jumps** to the end of the if-elif-else block without executing anything else. Thus, the actions under `else` are only executed if NONE of the previous conditions were True. If there is no ```else```, nothing will be done if all conditions evaluate to False.

The full syntax is like this, with as many `elif` blocks as you need:
```
if <condition A>:
    <do x>
elif <condition B>:
    <do y>
elif <condition C>:
    <do y>
else:
    <do z>
```

Say that you want to print something if the patient's age is greater than 40.

**Run the code cell below.**

````python
# in this example, patient_age is 45. What do you think should happen if you run this code?
patient_age = 45 
if patient_age > 40:
    print("Patient is more than 40 years old!")
````

Now change `patient_age` in the above code cell to 30 and run the code again. What happens when you run the code again? Do you see anything in the output?

You will not see anything in the output, because the condition ```patient_age > 40``` was False; therefore, the print statement was skipped. If we want to handle the case where the condition is False, we can add an `else` statement:

**Run the code cell below.**

````python
patient_age = 30
if patient_age > 40:
    print("Patient is more than 40 years old!")
else:
    print("Patient is 40 years old or younger!")
````

Now let's try a slightly more advanced exercise.

Say that you only want to include patients between the ages of 40 (inclusive) and 95 (not inclusive) in your study. You would write an expression like this, and patients that meet the criteria would return True when this expression is evaluated:
```
patient_age >= 40 and patient_age < 95
```

Let's break this down. The first part, `patient_age > 40`, is one logical condition. It uses the `>=` comparison operator, so we are evaluating if patient_age is greater than or equal to 40. ```patient_age < 95``` is another logical condition to test if patient_age is less than 95. Lastly, these two conditions are joined by `and`. This means that the condition to the left of `and ` (the first part) must be True **AND** the condition to the right (the second part) must be True for the entire expression to evaluate to True.

You may use brackets to explicitly specify the order that the conditions should be evaluated.
Run the code below to see an example of how conditional statements work. Do not worry about understanding all parts of the code, but pay attention to the if-else statement.

````python
patient1 = 35
patient2 = 40
patient3 = 67
patient4 = 100

def handle_patient(patient_age):
    if patient_age >= 40 and patient_age < 95:
        print(f"Patient age: {patient_age}, Include this patient")
    else:
        print(f"Patient age: {patient_age}, Exclude this patient")

handle_patient(patient1)
handle_patient(patient2)
handle_patient(patient3)
handle_patient(patient4)
````

### List indexing
In order to access elements from a list, we need a way to refer to elements by their location. We call this indexing.

 <span style="background-color: #AFEEEE">**Index**</span>: an integer number that refers to a location in a list.  

As mentioned earlier, lists are zero-indexed. Zero-indexing means the position of the elements in a list is numbered starting from 0. Because the first element is at position 0, the last element is at position len(list_name) - 1. In this list of 8 elements, the last element is at the 7th index.

<span style="background-color: #AFEEEE">**Zero-indexing**</span>: the position of the elements in a list is numbered starting from 0.

<img src="image-3.png" alt="Drawing" style="width: 400px;"/>


If we want to access elements near the end of the list, we can actually index into the list from the back instead of the front. Python allows negative indexing for this purpose. This means the last element is at index -1, the second last element is at index -2, and so on. This is handy if we want to access "the second last element", because we can simply use a negative index of -2 instead of finding out the length of the list using len() and calculating what the second last index would be.

<img src="image-4.png" alt="Drawing" style="width: 400px;"/>

To access a value in a list at index i, we would do list_name[i]. For example,

````python
patient[0]
````

### List slicing
Now we know how to use list indexing to access a single element. What if we want to get a subset of a list? We can get a subset of a list, from index i (inclusive) to j (NOT inclusive), using list slicing. The syntax is:

list_name[i:j]

Below is a list of daily high temperatures in Celsius for a week in August, starting from Sunday. If we want to know the daily highs for weekdays, we need a subset of this list. If we work out the indices for each day, it would look like this:

<img src="image-5.png" alt="Drawing" style="width: 800px;"/>

Weekdays start at index 1 and end at index 5. Remember that the second index in the slicing operation is NOT inclusive. Therefore, if we try temp[1:5], the operation will not include Friday. The second index needs to be 6:

````python
# Sunday, Monday, Tuesday, Wednesday, Thursday, Friday, Saturday
temp = [28, 26, 25, 27, 27, 29, 26]
temp[1:6]
````

There are some useful shorthands here. If we want a slice of the list from the beginning to some index j (i = 0, the start of the list), we can omit it: list_name[:j]

Conversely, if we want a slice of the list from some index i to the end of the list (j = len(list)-1), we can omit j: list_name[i:]

Say we want to get the first 4 elements of the patient record list. We can do:

````python
first_4 = patient[:4]
remaining = patient[4:]
first_4
````

When we do this, the new lists **first_4** and **remaining** are now separate lists, and each are zero-indexed.

![alt text](image-6.png)

````python
# these are the same values
print(patient[6], remaining[2])
````
